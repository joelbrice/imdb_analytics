```python

#Let's rank by the most expensive movies 

expensive_movies_queries ='''
SELECT
    movies.original_title,
    directors.name,
    movies.release_date,
    movies.budget,
    movies.revenue

FROM
    movies
JOIN
    directors
ON
    movies.director_id =directors.id
ORDER BY
    movies.budget DESC

'''

expensive_movies_data = cur.execute(expensive_movies_queries)

columns = [description[0] for description in cur.description]

expensive_movies_df = pd.DataFrame(expensive_movies_data, columns=columns)

expensive_movies_df.head(3)
```

```python
#Let's rank frim the most  revenue based movies:
best_revenues_movies_queries ='''
SELECT
    movies.original_title,
    directors.name,
    movies.release_date,
    movies.budget,
    movies.revenue

FROM
    movies
JOIN
    directors
ON
    movies.director_id =directors.id
ORDER BY
    movies.revenue DESC

'''

expensive_movies_data = cur.execute(best_revenues_movies_queries)

columns = [description[0] for description in cur.description]

expensive_movies_df = pd.DataFrame(expensive_movies_data, columns=columns)

expensive_movies_df.head(3)
```
